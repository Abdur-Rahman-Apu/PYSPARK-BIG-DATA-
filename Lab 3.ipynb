```python
import pyspark
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('Lab3').getOrCreate()
```

<h2>Create RDD</h2>

```python
myRdd=spark.sparkContext.parallelize([('Apu',1),('Yasin',2),('Apu',3),('Yasin',4),('Apu',5),('Yasin',6),('Apu',7),('Apu',8)])
```

```python
myRdd.collect()
```

```python
myRdd.take(2)
```

<h3>Read a text File</h3>

```python
textRDD=spark.sparkContext.textFile('lab3.txt')
```

```python
textRDD.collect()
```

```python
textRDD.getNumPartitions()
```

```python
def Func(line):
    line=line.lower()
    line=line.split()
    return line

splitRDD=textRDD.map(Func)
```

```python
splitRDD.collect()
```

```python
splitRDD.take(5)
```

```python
flatRDD=textRDD.flatMap(Func)
flatRDD.collect()
```

```python
flatRDD.take(5)
```

<h3>Work on flatmap output</h3>

```python
stopwords=['a','an','so','to','the','and']
```

```python
without_stopword=flatRDD.filter(lambda x: x not in stopwords)
```

```python
without_stopword.take(5)
```

```python
RDD2=flatRDD.filter(lambda x: x.startswith('d'))
```

```python
RDD2.collect()
```

```python
RDD2.distinct().collect()
```

<h3>Join</h3>

```python
a=spark.sparkContext.parallelize([('a',1),('b',2),('c',4)])
b=spark.sparkContext.parallelize([('a',2),('b',3)])
```

```python
c=a.join(b)
```

```python
c.collect()
```

<h3>Range</h3>

```python
n=spark.sparkContext.parallelize(range(0,10000))
```

```python
n.collect()
```

```python
n.reduce(lambda x,y:x+y)
```

<h3>Union,intersection,subtraction,cartesian</h3>

```python
a=spark.sparkContext.parallelize([1,2,3,4])
b=spark.sparkContext.parallelize([3,4,5,6])
```

```python
a.union(b).collect()
```

```python
a.intersection(b).collect()
```

```python
a.subtract(b).collect()
```

```python
a.cartesian(b).collect()
```
